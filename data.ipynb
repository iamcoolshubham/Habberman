```python
from matplotlib.colors import ListedColormap
def knn_comparison(dataframe, n_neighbors = 1):
    '''
    This function finds k-NN and plots the data.
    '''
    X = dataframe[1:, :2]
    y = dataframe[1:,2] 
    
    
    # grid cell size
    h = .02
    cmap_light = ListedColormap(['#FFAAAA',  '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#0000FF'])


    # the core classifier: k-NN
    clf = neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    # we create a mesh grid (x_min,y_min) to (x_max y_max) with 0.02 grid spaces
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    # we predict the value (either 0 or 1) of each element in the grid
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
    # xx.ravel() will give a flatten array

    # np.c_ : Translates slice objects to concatenation along the second axis.
    # > np.c_[np.array([1,2,3]), np.array([4,5,6])]
    # > array([[1, 4],
    #          [2, 5],
    #          [3, 6]])   (source: np.c_ documentation)
    
    
    
    
    # convert the out back to the xx shape (we need it to plot the decission boundry)
    Z = Z.reshape(xx.shape)
    
    
    # pcolormesh will plot the (xx,yy) grid with colors according to the values of Z
    # it looks like decision boundry
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
   
    # scatter plot of with given points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
    
    #defining scale on both axises
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
   
    # set the title
    plt.title('K value = '+str(n_neighbors))
    

    plt.show()
```

```python
data = np.genfromtxt('new.csv', delimiter=',')
knn_comparison(data, 1)
knn_comparison(data, 5)
knn_comparison(data,15)
knn_comparison(data,30)
```

# Predict that status of patient

```python
a = data[1:, :2]
b= data[1:,2]
k=int(input("Enter the K value for nearest neighbour"))
clf = neighbors.KNeighborsClassifier(k)
clf.fit(a,b)
x1=float(input('Enter the 1st component of the patient'))
x2=float(input('Enter the 2nd component of the patient'))
z=clf.predict([[x1,x2]])
if(z==1):
    print("The patient is most likely survive more than 5 yrs")
else:
    print("The patient is most likely to survive less than 5 yrs")
```

# Predict the survival and death probablity of patient


```python
p=list()
p=clf.predict_proba([[x1,x2]])
print("The probablity that patient will survive more than 5 yrs is {} %".format(p[0][0]*100))
print("The probablity that patient will survive less than 5 yrs is {} %".format(p[0][1]*100))

```
